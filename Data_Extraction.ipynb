# Visualizations

```python
!pip install mplfinance
```

```python
import matplotlib.pyplot as plt
from mplfinance import *
from mplfinance.original_flavor import candlestick_ohlc
```

```python
from plotly.graph_objs import Candlestick
import plotly
```

```python
df = train[["Open", "High", "Low", "Close"]]
trace = Candlestick(
    x=df.index,
    open=df["Open"],
    high=df["High"],
    low=df["Low"],
    close=df["Close"],
)

ticker='NIFTY 50'
# Create and display plotly figure
fig = {
    "data": [trace],
    "layout": {"title": f"{ticker} Candlestick Chart"},
}

plotly.offline.plot(fig, filename="candlestick.html")
```

```python
df = nifty_historical_data[["Open", "High", "Low", "Close"]]
trace = Candlestick(
    x=df.index,
    open=df["Open"],
    high=df["High"],
    low=df["Low"],
    close=df["Close"],
)

ticker='NIFTY 50'
# Create and display plotly figure
fig = {
    "data": [trace],
    "layout": {"title": f"{ticker} Candlestick Chart"},
}

plotly.offline.plot(fig, filename="candlestick_5m.html")
```
